```python
import numpy as np
import random

def create_8_queen_board():
    # Create an 8x8 board initialized with '.' (no queens placed)
    queen_board = np.full((8, 8), '.')  # Corrected shape from 8,8 to (8, 8)

    # Place queens ('Q') on the board
    for row in range(8):
        col = random.randint(0, 7)  # Random column for each row
        print(row,col)
        queen_board[row][col] = 'Q'

    return queen_board

def board_chromosome(board):
    # Find positions of queens on the board
    queen_positions = np.argwhere(board == 'Q')
    col = queen_positions[:, 1]  # Extract column indices of queens
    return col

def fitness_fun_numpy(board):
    # Get positions of the queens
    queen_positions = np.argwhere(board == 'Q')
    print("Queen Positions:",queen_positions)
    num_queens = len(queen_positions)  # Total number of queens
    print("Number of queens:", num_queens)

    # Extract rows and columns
    row = queen_positions[:, 0]
    col = queen_positions[:, 1]
    print("Rows:", row)
    print("Columns:", col)
    print("Diagnols:",row-col)
    print("Antidiagnol:",row+col)


    # Count attacking pairs
    col_attack = num_queens - len(np.unique(col))  # Count columns with multiple queens
    diag_attack = num_queens - len(np.unique(row - col))  # Main diagonal
    anti_diag_attack = num_queens - len(np.unique(row + col))  # Anti diagonal

    # Total attacks
    total_attacks = col_attack + diag_attack + anti_diag_attack

    # Fitness is the number of non-attacking queens
    fitness = num_queens - total_attacks
    return fitness

boards = []
chromosomes = []
fitness_values = []

# Create and evaluate multiple boards
for i in range(4):  # Create four boards
    board = create_8_queen_board()
    print(f"Board {i+1}:\n{board}\n")
    chromosome = board_chromosome(board)
    print(f"Chromosome C{i+1}:", chromosome)
    fitness_value = fitness_fun_numpy(board)  # Use a different variable name for fitness
    print(f"Fitness of B{i+1}:", fitness_value)
    print("-------------------")

    # Append the board, chromosome, and fitness to their respective lists
    boards.append(board)  # Append the current board
    chromosomes.append(chromosome)  # Append the current chromosome
    fitness_values.append(fitness_value)  # Append the current fitness value

# Print the lists of boards, chromosomes, and fitness values
print("Boards List:")
for b in boards:
    print(b)
print("Chromosomes List:", chromosomes)
print("Fitness Values List:", fitness_values)



```

```python
def top_two_chormosomes(fitness,boards,chromosomes):
  combined_list=list(zip(fitness_values,boards,chromosomes)) #combine 3 lists into single list in which there are 3 tuples each contain
  #fitness ,board and chromosome of boards
  sorted_list=sorted(combined_list,key=lambda x: x[0]) ##arrange the lists based on
  #first parameter value i.e fitness function in ascending order i.e low comes first.
  top_two_chromosomes=sorted_list[-2:]
  return top_two_chromosomes

top_two=top_two_chormosomes(fitness_values,boards,chromosomes)
print("Top_two:",top_two)

parent1=top_two[-1][2]          # Chromosome of the first top board
parent2=top_two[-2][2]      # Chromosome of the second top board
print("Parent1:",parent1)
print("Parent2:",parent2)

def single_point_crossover(parent1,parent2):
  #randomly chooses crossover
  ##crossover_point=np.random.randint(1,7)
  #crossover_point=np.random.randint(1,len(parent1)-1)  #1 to 7 ..1,8-1=1,7
  crossover_point=4
  print("Crossover Point:",crossover_point)

  offspring1=np.concatenate((parent1[:crossover_point],parent2[crossover_point:]))
  offspring2=np.concatenate((parent2[:crossover_point],parent1[crossover_point:]))
  return offspring1,offspring2

offspring1,offspring2=single_point_crossover(parent1,parent2)

print("Offspring1:",offspring1)
print("Offspring2:",offspring2)
```

```python
def mutation(offspring1,offspring2):
  mutation_point=np.random.randint(0,7)
  print("Mutation Point:",mutation_point)
  offspring1[mutation_point]=np.random.randint(0,7)
  offspring2[mutation_point]=np.random.randint(0,7)
  print(offspring1[mutation_point])
  print(offspring2[mutation_point])
  return offspring1,offspring2

offspring1,offspring2=mutation(offspring1,offspring2)
print("Offspring1:",offspring1)
print("Offspring2:",offspring2)
```

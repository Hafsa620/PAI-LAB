- Remove special characters anf extra spaces

```python
import re
def remove_special_char(text):
  text=re.sub('[^a-zA-Z0-9\s]','',text)
  return text
def remove_extra_space(text):
  text=re.sub('\s+',' ',text)
  return text
```

```python
data['v2']=data['v2'].apply(remove_special_char)
data['v2']=data['v2'].apply(remove_extra_space)
```

```python
data['v2'][2]
```

```python
data
```

- Stopwords removal

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords
stw=set(stopwords.words('english'))
```

```python
def stopword_removal(text):
  text=[word for word in text.split() if word not in stw]
  return ' '.join(text)
```

```python
data['v2']=data['v2'].apply(stopword_removal)
```

```python
data
```

- Tokenization

```python
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import word_tokenize,sent_tokenize
```

```python
data['v2']=data['v2'].apply(word_tokenize)
```

```python
data
```

-Lemmatization

```python
nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatization(text):
  text=[lemmatizer.lemmatize(word) for word in text]
  return ' '.join(text)
```

```python
data['v2']=data['v2'].apply(lemmatization)
data
```

- Remove irrelevant columns

```python
data.drop(['Unnamed: 2','Unnamed: 3','Unnamed: 4'],axis=1,inplace=True)
```

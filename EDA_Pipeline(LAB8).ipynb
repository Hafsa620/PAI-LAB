```python
import pandas as pd
from sklearn.compose import ColumnTransformer
from sklearn.impute import SimpleImputer
from sklearn.preprocessing  import OneHotEncoder, FunctionTransformer,MinMaxScaler
from sklearn.pipeline import Pipeline
```

```python
data=pd.read_csv('Titanic.csv')
```

Column Transformer

```python
preprocessor=ColumnTransformer(transformers=[('age_imputer',SimpleImputer(strategy='mean'),['Age']),
                                             ('ohe',OneHotEncoder(),['Sex','Embarked'])],remainder='passthrough')

transformed_data=preprocessor.fit_transform(data)
print(f'Transformed Data:{transformed_data}')

print(f'Transformed Data:{pd.DataFrame(transformed_data)}')

transformed_data.shape
```

Function transformer

```python
def impute_embarked(X):
  X['Embarked']=X['Embarked'].fillna(X['Embarked'].mode()[0])
  return X

preprocessor=ColumnTransformer(transformers=[('age_imputer',SimpleImputer(strategy='mean'),['Age']),
                                             ('ohe',OneHotEncoder(),['Sex','Embarked']),('embarked_imputer',FunctionTransformer(impute_embarked),['Embarked'])],remainder='passthrough')

transformed_data=preprocessor.fit_transform(data)
print(f'Transformed Data:{transformed_data}')

print(f'Transformed Data:{pd.DataFrame(transformed_data)}')

transformed_data.shape
```

pipeline

```python
def impute_embarked(X):
  X['Embarked']=X['Embarked'].fillna(X['Embarked'].mode()[0])
  return X

preprocessor=ColumnTransformer(transformers=[('age_imputer',SimpleImputer(strategy='mean'),['Age']),
                                             ('embarked_encoder',Pipeline(steps=[('embarked_imputer',FunctionTransformer(impute_embarked)),('ohe',OneHotEncoder())]),['Embarked']),
                                              ('onehot',OneHotEncoder(),['Sex'])],remainder='passthrough')

transformed_data=preprocessor.fit_transform(data)
print(f'Transformed Data:{transformed_data}')

print(f'Transformed Data:{pd.DataFrame(transformed_data)}')

transformed_data.shape
```

Family size through Function transformer

```python
def family_size(data):
  data['FamilySize']=data['SibSp']+data['Parch']
  return data

def impute_embarked(X):
  X['Embarked']=X['Embarked'].fillna(X['Embarked'].mode()[0])
  return X

preprocessor=ColumnTransformer(transformers=[('age_encoder',Pipeline(steps=[('age_imputer',SimpleImputer(strategy='mean')),
                                             ('age_scaler',MinMaxScaler())]),['Age']),
                                             ('embarked_encoder',Pipeline(steps=[('embarked_imputer',FunctionTransformer(impute_embarked)),('ohe',OneHotEncoder())]),['Embarked']),
                                              ('onehot',OneHotEncoder(),['Sex']),('fare_scaler',MinMaxScaler(),['Fare']),('Family_size',FunctionTransformer(family_size),['SibSp','Parch'])],remainder='passthrough')

transformed_data=preprocessor.fit_transform(data)
#print(f'Transformed Data:{transformed_data}')
print(f'Original Data:{data}')


print(f'Transformed Data:{pd.DataFrame(transformed_data)}')

transformed_data.shape
```

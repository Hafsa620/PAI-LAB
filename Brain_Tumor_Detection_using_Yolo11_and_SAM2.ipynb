# Extracting bounding boxes from yolo detected images and then feed those coordinated to Meta model i.e SAM2(Segment anything model) for segmentation on extracted coordinates

```python
from ultralytics import YOLO, SAM
import os, torch

# Load models
yolo_model = YOLO("/content/runs/detect/train/weights/best.pt")
sam_model = SAM("sam2_b.pt")

# Folder with already predicted images
Detection_results = "/content/runs/detect/predict3"

# Loop through all images in that folder
for filename in os.listdir(Detection_results):
    if filename.endswith(('.jpg', '.png')):
        image_path = os.path.join(Detection_results, filename)

        # Run YOLO again only to extract boxes (suppress label drawing)
        results = yolo_model.predict(
            source=image_path,
            save=False,             # Don't save annotated image
            verbose=False,
            show_labels=False       # ✅ Disable class labels
        )

        for result in results:
            class_ids = result.boxes.cls.int().tolist()
            if len(class_ids):
                boxes = result.boxes.xyxy.cpu()

                # Run SAM to generate segmentation mask
                sam_results = sam_model.predict(
                    source=image_path,
                    bboxes=boxes,
                    verbose=False,
                    save=True,        # Save segmented mask
                    show_labels=False,  # ✅ Disable labels on output
                    device='cuda' if torch.cuda.is_available() else 'cpu'
                )
```
